```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
```

```python
translator = pipeline('translation', model=model, tokenizer=tokenizer, src_lang="ckb_Arab", tgt_lang='eng_Latn', max_length = 400)
```

```python
translator("سڵاو هاوڕێیان، پێتان چۆنە ڕۆژی هەینی بچین بۆ مەلە؟")
```

```python
translator("پلەکانی گەرما ڕۆژ بە ڕۆژ بەرەو بەرزبونەوە دەچن.")
```
